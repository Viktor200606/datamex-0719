```python
# Tu codigo aqui
import pymongo
from pymongo import MongoClient
import folium
from folium import plugins
import pandas as pd
from bson.json_util import loads
from folium.plugins import HeatMap
```

```python
def g_point(punto):
    return { "type":"Point", "coordinates":[punto[0],punto[1]] }
```

```python
startups1 = pd.read_json('startups.json')
startups1.head()
```

```python
startups1['location'] = startups1.apply(lambda x: [x.latitude, x.longitude], axis=1)
startups1['Geolocation'] = startups1.apply(lambda x: { "type":"Point", "coordinates":[x.longitude, x.latitude]}, axis=1)

startups1.head()
```

```python
def findNear(geopoint={'type': 'Point', 'coordinates': [-122.1612868, 37.4440981]}, radio_max_meters=3000):
    return db.p_goloc.find({
                          'Geolocation': {
                                           "$near": { "$geometry": geopoint,"$maxDistance": radio_max_meters,}
                                          }
                        })

radio_max_meters = 3000

#for oficina in startups1.location
p_startup=[]
maximo=0

for oficina in list(colec.find( {'founded_year': {'$gte': 2008}}, {'Geolocation':1, '_id':0})):
    num_offices = findNear(oficina['Geolocation']).count()
    if maximo < num_offices:
        maximo = num_offices
        p_startup.append(oficina['Geolocation']['coordinates'])
        
print('La nueva oficina puede estar cerca de los puntos ', p_startup,\
      'por que tienen ', maximo, 'oficinas en un radio de', radio_max_meters, 'm')
```

```python
def generateBaseMap(default_location=[37.775087, -122.415199], default_zoom_start=12):
    return folium.Map(location=default_location, control_scale=True, zoom_start=default_zoom_start)

base_map = generateBaseMap()
HeatMap(data=startups1.location, radius=8, max_zoom=13).add_to(base_map)

for i in p_startup:
    folium.Marker([i[1],i[0]], icon=folium.Icon(color='red')).add_to(base_map);
display (base_map)
```
