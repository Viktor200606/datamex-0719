El primer paso es seleccionar lo datos desde MongoDB, pasarlos a un dataframe, limpiarlos y guardar las seleccion de datos en un archivo json para volver a llevarlos a MongoDB.

```python
import pandas as pd
import numpy as np
```

```python
# Tu codigo aqui
from pymongo import MongoClient
client=MongoClient()
db=client.companies
db.companies
```

```python
#convert entire collection to Pandas dataframe
companias = pd.DataFrame(list(db.companies.find()))
companias
```

```python
pd.isnull(companias).sum().sort_values(ascending=False)
```

```python
companias_d=companias.dropna(1)
companias_d
```

```python
pd.isnull(companias_d).sum().sort_values(ascending=False)
```
