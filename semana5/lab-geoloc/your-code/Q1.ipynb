El primer paso es seleccionar lo datos desde MongoDB, pasarlos a un dataframe, limpiarlos y guardar las seleccion de datos en un archivo json para volver a llevarlos a MongoDB.

```python
import pandas as pd
import numpy as np
import json
from pandas.io.json import json_normalize
```

```python
# Tu codigo aqui
from pymongo import MongoClient
client=MongoClient()
db=client.companies
db.companies
```

```python
columnas = pd.DataFrame(list(db.companies.find()))
columnas
```

```python
#companies1=db.companies.find( { '_id': {'$ne': []} } )
#for i in columnas.columns[1:]:
#    companies1=db.companies1.find( { '$or': [ { i: {'$ne': []} }, {i: None} ] } )
#list(companies1[:20])
```

```python
companies1 = db.companies.find({ '$and': [{'number_of_employees': {'$lt': 1000}},\
                                          {'number_of_employees': {'$gte': 1}},\
                                          
                                          {'founded_year': {'$gte': 2008}}]},\
                               {'number_of_employees':1,\
                                'founded_year':1, '_id':0, 'offices':1}).sort('number_of_employees', -1)
companias_clean=list(companies1)
list(companias_clean)
```

```python
startup = pd.concat([pd.DataFrame(companias_clean)[['founded_year', 'number_of_employees']], pd.DataFrame(json_normalize(companias_clean, 'offices'))[['longitude','latitude']]],axis=1)
startup.head()
```

```python
companias_clean=pd.DataFrame(startup)
print(companias_clean.shape)
companias_clean.head()
```

```python
pd.isnull(companias_clean).sum().sort_values(ascending=False)
```

```python
companias_d=companias_clean.dropna()
print(companias_d.shape)
companias_d.head()
```

```python
pd.isnull(companias_d).sum().sort_values(ascending=False)
```

```python
companias_d.head()
```

```python
companias_d.to_json('startups.json')
#se guarda data limpia y se guarda en rama lab-geoloc
```
